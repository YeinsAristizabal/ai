Data preparation and cleaning

Identification of Missing Values

```python
print(df.isnull().sum())
```

Imputation of data using the mean in the Income column 

```python
df['Income'] = df['Income'].fillna(int(df['Income'].mean()))
print(df.isnull().sum())
```

Conversion of variables

```python
# Assign ordinal values to the Education variable
education_mapping = {
    'Basic': 1,
    'Graduation': 2,
    'Master': 3,
    'PhD': 4,
    '2n Cycle': 5
}
df['Education'] = df['Education'].map(education_mapping)
df['Education'].head(5)
```

```python
# Calculate the frequency of each category
marital_freq = df['Marital_Status'].value_counts(normalize=True)

# Mapping frequencies to categories
df['Marital_Status'] = df['Marital_Status'].map(marital_freq)
df['Marital_Status'].head(5)
```

Preparation of variables

Recency Variable

```python
# Verify extreme values
print(df['Recency'].describe())
```

```python
# Standardization
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df['Recency_scaled'] = scaler.fit_transform(df[['Recency']])
df['Recency_scaled'].head(5)
```

Income Variable

```python
# Verify extreme values
print(df['Income'].describe())
```

```python
# Apply filter: We asume that value is an outlier
df = df[df['Income'] < 666666]
print(df['Income'].describe())
```

```python
# Standarization
df['Income_scaled'] = scaler.fit_transform(df[['Income']])
df['Income_scaled'].head(5)
```

Total Expenses Variable

```python
# Creating a Total Expenses column
df['Total_Spending'] = df[['MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']].sum(axis=1)

# Verify extreme values
print(df['Total_Spending'].describe())
```

```python
# Standardization
df['Total_Spending_scaled'] = scaler.fit_transform(df[['Total_Spending']])
df['Total_Spending_scaled'].head(5)
```

Number of purchases variable

```python
# Creating a Number of Purchases column
df['Total_Purchases'] = df[['NumWebPurchases', 'NumCatalogPurchases', 'NumStorePurchases']].sum(axis=1)

# Verify extreme values
print(df['Total_Purchases'].describe())
```

```python
# Standardization
df['Total_Purchases_scaled'] = scaler.fit_transform(df[['Total_Purchases']])
df['Total_Purchases_scaled'].head(5)
```

Response to campaigns variable

```python
# Creating a Response to campaigns column
campaign_columns = ['AcceptedCmp1', 'AcceptedCmp2', 'AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5', 'Response']
df['Total_Campaign_Response'] = df[campaign_columns].sum(axis=1)
print(df['Total_Campaign_Response'].describe())
```

```python
df['Total_Campaign_Response'].head(5)
```

Predictive model building

Dataset Consolidation

```python
# Relevant variables
variables_modelo = [
    'Recency_scaled',
    'Income_scaled',
    'Total_Spending_scaled',
    'Total_Purchases_scaled',
    'Total_Campaign_Response'
]

# Cconsolidated dataset
final_dataset = df[variables_modelo]

final_dataset.head()
```

Model training

```python
# Train the K-Means model with the optimal number of clusters.
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

optimal_k = 2  # Assuming that we want to segment customers into two groups
kmeans_final = KMeans(n_clusters=optimal_k, random_state=42)
final_dataset['Cluster'] = kmeans_final.fit_predict(final_dataset)
```

```python
final_dataset
```

```python
# Analysis of the size of each cluster
cluster_sizes = final_dataset['Cluster'].value_counts()
print("Cluster size:")
print(cluster_sizes)
```

Denormalization

```python
df_denormalized = scaler.inverse_transform(final_dataset[['Recency_scaled','Income_scaled','Total_Spending_scaled','Total_Purchases_scaled']])
df_denormalized = pd.DataFrame(df_denormalized, columns=['Recency_scaled','Income_scaled','Total_Spending_scaled','Total_Purchases_scaled'])
df_denormalized['Total_Campaign_Response'] = final_dataset['Total_Campaign_Response']
df_denormalized['Cluster'] = final_dataset['Cluster']
df_denormalized
```

Cluster visualization

```python
import seaborn as sns

# 2D visualization of the clusters (using Recency_scaled & Income_scaled variables)
sns.scatterplot(
    x=df_denormalized['Recency_scaled'],
    y=df_denormalized['Income_scaled'],
    hue=df_denormalized['Cluster'],
    palette='viridis'
)
plt.title('Cluster visualization')
plt.xlabel('Recency')
plt.ylabel('Income')
plt.legend(title='Cluster')
plt.show()
```

```python
# 2D visualization of the clusters (using Total_Spending_scaled & Income_scaled variables)
sns.scatterplot(
    x=df_denormalized['Total_Spending_scaled'],
    y=df_denormalized['Income_scaled'],
    hue=df_denormalized['Cluster'],
    palette='viridis'
)
plt.title('Cluster visualization')
plt.xlabel('Total Spending')
plt.ylabel('Income')
plt.legend(title='Cluster')
plt.show()
```

```python
# 2D visualization of the clusters (using Income_scaled & Total_Purchases_scaled variables)
sns.scatterplot(
    x=df_denormalized['Income_scaled'],
    y=df_denormalized['Total_Purchases_scaled'],
    hue=df_denormalized['Cluster'],
    palette='viridis'
)
plt.title('Cluster visualization')
plt.xlabel('Income')
plt.ylabel('Total Purchases')
plt.legend(title='Cluster')
plt.show()
```
